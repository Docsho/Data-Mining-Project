```python
# import library
import numpy as np
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'fr-FR,fr;q=0.9,en-US;q=0.8,en;q=0.7',
}

data1 = []

for i in range(1, 50):
    if len(data1) >= 1000:
        break

    url = 'https://www.jumia.ci/peripheriques-logiciels-accessoires/?q=telephone#catalog-listing'  # Adjust this URL
    
    response = requests.get(url, headers=headers)
    response = response.content

    soup = BeautifulSoup(response, 'html.parser')

    product_cards = soup.select("article.prd")

    for card in product_cards:

        # product name
        name_tag = card.select_one("h3.name")
        name = name_tag.get_text(strip=True) if name_tag else "N/A"

        # product price
        price_tag = card.select_one("div.prc")
        price = price_tag.get_text(strip=True) if price_tag else "N/A"


        # Image 
        img_tag = card.select_one("img")
        image_url = ""
        if img_tag:
            image_url = img_tag.get("data-src") or img_tag.get("src") or "N/A"

        data1.append({
            'name': name,
            'price': price,
            'image_url': image_url
        })

print(data1)
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df1['name'] = df1['name'].replace("N/A", np.nan)
df1 = df1.dropna(subset=['name']).reset_index(drop=True)
df1
```

```python
df1['name'] = df1['name'].str.lower()
df1
```

```python
import re
df1['price'] = df1['price'].apply(
    lambda p: max([int(x.replace(',', '').replace(',', '')) 
                   for x in re.findall(r'\d[\d.,]*', str(p))])
)
df1
```

```python
df1 = df1.dropna(subset=['name', 'price', 'image_url']).reset_index(drop=True)
df1
```

```python

from mlxtend.frequent_patterns import apriori, association_rules

#  Convert DataFrame en booleen
df_basket_bool = df_basket.astype(bool)

# Apply algorithme Apriori algo
freq_items = apriori(df_basket_bool, min_support=0.01, use_colnames=True)

print("Itemsets fréquents :")
print(freq_items)

# --- Étape 3 : Générer les règles d'association ---
rules = association_rules(freq_items, metric="confidence", min_threshold=0.07)
print("\nColonnes disponibles dans rules :")
print(rules.columns)
print("\nExemple de règles :")
print(rules.head())

#recommandation fonction base on rules
def recommend_products(product, rules_df, top_n=2):
    recommendations = []

    # Chercher les règles où le produit est dans les antécédents
    matching_rules = rules_df[rules_df['antecedents'].apply(lambda x: product in x)]
    
    # Vérifier si des règles existent
    if matching_rules.empty:
        return recommendations

    # Trier par confiance décroissante
    matching_rules = matching_rules.sort_values(by='confidence', ascending=False)

    # Ajouter les produits des conséquents aux recommandations
    for _, row in matching_rules.iterrows():
        for item in row['consequents']:
            if item not in recommendations:
                recommendations.append(item)
                if len(recommendations) >= top_n:
                    break
        if len(recommendations) >= top_n:
            break

    return recommendations

# --- Étape 5 : Exemple d'utilisation ---
product_to_recommend = 'lecteur de cartes otg 3-en-1 usb'
recommendations = recommend_products(product_to_recommend, rules)
print(f"\nRecommandations pour '{product_to_recommend}' : {recommendations}")
print(recommendations)
```

```python
# --- STEP 1 : Create mapping from product names to their IDs ---
# This allows us to convert Apriori antecedents/consequents (names)
# into numerical IDs used by the website frontend.
name_to_id = dict(zip(products_df["name"], products_df["id"]))

rows = []   # This will store final recommendation pairs


# --- STEP 2 : Convert Apriori rules to recommendation pairs ---
# For every rule:
#   antecedent product A  ---> consequent product B
# We store:   product_id = ID(A),   rec_id = ID(B)
# These pairs are exactly what the website uses.
for _, row in rules.iterrows():

    # A rule might contain several products in antecedents or consequents
    for antecedent in row["antecedents"]:
        for consequent in row["consequents"]:

            # Only save the rule if both products exist in our dataset
            if antecedent in name_to_id and consequent in name_to_id:

                rows.append({
                    "product_id": name_to_id[antecedent],  # the product we clicked
                    "rec_id": name_to_id[consequent]       # the recommended product
                })


# --- STEP 3 : Convert to DataFrame and remove duplicates ---
recommend_df = pd.DataFrame(rows).drop_duplicates()

# --- STEP 4 : Save final recommendation mapping ---
# This CSV will be read by details_jumia.js in the website.
recommend_df.to_csv("jumia_recommendations.csv", index=False)

# Preview
recommend_df.head()
```
