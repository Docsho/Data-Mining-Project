```python
import requests
from bs4 import BeautifulSoup

data = []

# for loop that extract data for each page
for i in range(1, 119):
    if len(data) >= 1000:
        break

    # Link of the pages
    url = f"https://ci.coinafrique.com/categorie/voitures?page={i}"

    # Connection to the link
    response = requests.get(url)
    response = response.content

    # Convert the code into a HTML code
    soup = BeautifulSoup(response, 'html.parser')

    # Find the div class of the HTML code
    div = soup.find('div')

    # Find all div of the class = col s6 m4 l3
    vehicules = soup.find_all('div', class_='col s6 m4 l3')

    # Find the element of vehicules specifically
    for vehicule in vehicules:

        # find the Price of each vehicule
        price = vehicule.find('p', class_='ad__card-price')
        price = price.text.strip() if price else 'N/A'

        # find the description of each vehicule
        car_name = vehicule.find('p', class_='ad__card-description')
        car_name = car_name.text.strip() if car_name else 'N/A'

        # find the location of each vehicule
        location = vehicule.find('p', class_='ad__card-location')
        location = location.text.strip() if location else 'N/A'

        # image extraction
        img_tag = vehicule.find('img')
        image_url = img_tag.get('src') or img_tag.get('data-src') if img_tag else "N/A"

        data.append({
            'Price': price,
            'Location': location,
            'car_name': car_name,
            'Image_URL': image_url
        })

print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
# Load your cars CSV
df = pd.read_csv("cars.csv")

# Your recommender function
def recommendercar_sys(car_name, df, n=10,
                       price_weight=0.4, year_weight=0.4, brand_model_weight=0.5):

    car_name_lower = car_name.lower()

    matches = df[df["car_name"].str.lower().str.contains(car_name_lower)]
    if matches.empty:
        return pd.DataFrame()

    idx = matches.index[0]

    df["Price"] = df["Price"].astype(float)
    df["year"] = df["year"].astype(float)

    price_scores = 1 / (1 + abs(df['Price'] - df.loc[idx, 'Price']))
    price_scores = price_scores / price_scores.max()

    year_scores = 1 / (1 + abs(df['year'] - df.loc[idx, 'year']))
    year_scores = year_scores / year_scores.max()

    brand_model_bonus = (
        (df['brand'] == df.loc[idx, 'brand']) &
        (df['model'] == df.loc[idx, 'model'])
    ).astype(float)

    final_scores = (
        price_weight * price_scores +
        year_weight * year_scores +
        brand_model_weight * brand_model_bonus
    )

    final_scores[idx] = -1

    top_idx = final_scores.sort_values(ascending=False).iloc[:n].index

    results = df.loc[top_idx][[
        "car_name", "brand", "model", "Price",
        "year", "location", "Image_URL"
    ]]

    results["id"] = df.loc[top_idx].index
    return results


# Generate full recommendation table
rows = []

for i, row in df.iterrows():
    recs = recommendercar_sys(row["car_name"], df, n=10)
    for _, r in recs.iterrows():
        rows.append({
            "car_id": i,
            "rec_id": r["id"]
        })

recommend_df = pd.DataFrame(rows)

recommend_df.head()
```
